```python
import pandas as pd
```

```python
df = pd.read_csv("Salaries.csv")
df.set_index("Id")
```

```python
df.describe(include="all")
```

```python
def find_police(x):
    return "POLICE" in x

df["isPolice"] = df["JobTitle"].apply(find_police)
df["isPolice"].sum()





```

```python
def find_fire(x):
    return "FIRE" in x

df["isFire"] = df["JobTitle"].apply(find_fire)
df["isFire"].sum()
```

```python
ratio_fire_police = (df["isFire"].sum()) / (df["isPolice"].sum())
print(ratio_fire_police)
```

```python
police_df = df[df['JobTitle'].str.contains("POLICE")]
police_df
```

```python
fire_df = df[df['JobTitle'].str.contains("FIRE")]
fire_df
```
